```python
#2)Program to plot(x,y)
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')
#make data
x=np.linspace(0,10,100)
y=4+2*np.sin(2*x)
fig,ax=plt.subplots()
ax.plot(x,y,linewidth=2.0)
ax.set(xlim=(0,8),xticks=np.arrange(1,8),ylim=(0,8),yticks=np.arrange(1,8))
plt.show()
```
