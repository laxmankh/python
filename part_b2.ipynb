```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from scipy.stats import norm
# Generate random test results for a class (replace with your actual data)
np.random.seed(0)
test_results = np.random.normal(70, 10, 1000)  # Mean: 70, Standard Deviation: 10


plt.figure(figsize=(10, 6))
plt.hist(test_results, bins=30, density=True, color='skyblue')
plt.title('Test Results Distribution')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')
plt.grid(True)

# Add a normal distribution curve


mu, std = np.mean(test_results), np.std(test_results)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')



skewness = skew(test_results)
kurt = kurtosis(test_results)

print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurt}')


plt.legend()
plt.show()
```
